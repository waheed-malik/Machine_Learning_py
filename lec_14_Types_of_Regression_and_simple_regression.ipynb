### Regression

- Used to predict continuous values.
- The output is a real number (e.g., predicting house prices, temperatures, or salaries).
- Example: Predicting the price of a car based on its age, mileage, and model.

### Classification:

- Used to predict discrete categories or classes.
- The output is a class label (e.g., predicting whether an email is spam or not spam).
- Example: Classifying images as either cats or dogs.

```python
from sklearn.model_selection import train_test_split
```

```python
#x_train,x_test,y_train,y_test = train_test_split()
x = df.iloc[:,:-1]
y = df["Salary"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=10)
```

#### Purpose of random_state:

    - When you split your data into training and testing sets, the splitting is done randomly. If you   dont specify a random_state, the function will split the data differently each time you run the code.

    - by setting random_state to a specific integer (like 10 in your example), you are seeding the random number generator. This ensures that you always get the same split every time you run the code, making your results reproducible.

```python
from sklearn.linear_model   import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.predict([[2.3]])
```

### Least Square method:
   
   - a mathematical technique used to find the best-fitting line or model for a set of data points by minimizing the sum of the squared differences (errors) 
   between the actual data points and the predicted values. This method is widely used in regression 
   analysis to fit a line or curve to data and make predictions.
