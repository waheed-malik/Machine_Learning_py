### Let's Practice 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv("students_placement.csv")
ds.head(3)
```

```python
x = ds[["cgpa", 'iq', 'score']]
y = ds["placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size=.20, random_state=70)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
lr.score(x_test,y_test)*100
```

```python
from sklearn.metrics import confusion_matrix, recall_score, precision_score,f1_score
```

```python
cmr = confusion_matrix(y_test,y_pred)
```

```python
#ds["placed"].value_counts()
```

```python
sns.heatmap(data=cmr, annot=True)
plt.show()
```

```python
precision_score(y_test,y_pred)*100
```

```python
recall_score(y_test,y_pred)*100
```

```python
f1_score(y_test,y_pred)*100
```
