
A discrete variable represents countable values, often whole numbers, where there are no intermediate values (e.g., number of people, number of cars).

A continuous variable represents data that can take any value within a range, including decimals (e.g., height, weight, temperature).

Age and Salary:

    Age: 
    
    While in your dataset the values are whole numbers (15, 20, 25, etc.), age is typically considered continuous because, in reality, age can take any value (e.g., 25.5 years). The fact that you only have whole numbers doesn't make it discrete.


    Salary: 
    
    Similarly, salary can theoretically take any value, including decimals (e.g., $25,000.50). Even though your dataset has whole numbers, salary is still considered a continuous variable because it can vary smoothly.

"Inconsistent data" 

refers to data that is not uniform, reliable, or logically coherent across a system or database. This can happen when:

    1.1 Conflicting values: 
    
        The same piece of information is recorded differently in multiple places (e.g., a customer’s address being different in two systems).

    1.2 Incomplete data: 
        Some records are missing key information.

    1.3 Mismatched formats: 
    
        Data is stored in different formats (e.g., dates being stored as "MM/DD/YYYY" in one place and "DD/MM/YYYY" in another).

    1.4 Duplicate entries: 
    
        The same data is repeated multiple times in different forms.

Inconsistent data can lead to errors, unreliable analysis, and poor decision-making.
