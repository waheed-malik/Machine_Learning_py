In summary:

MSE and RMSE are measures of average squared errors, with RMSE being in the same scale as the target variable.
MAE is a measure of average absolute errors, providing a more interpretable metric than MSE.
R-squared measures the goodness of fit of the model, indicating how well the model's predictions explain the variance in the target variable.

Choosing the appropriate evaluation metric depends on the specific requirements of the problem, such as the desire for interpretability (MAE), sensitivity to outliers (MSE, RMSE), or overall model performance (R-squared).

MSE effected by outliers

RMSE not affected

MAE not affected.
