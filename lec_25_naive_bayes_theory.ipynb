```python
import matplotlib.pyplot as plt

# Probability of success
p = 0.7

# Values of X (0 or 1)
x = [0, 1]
# Probability values for each outcome
probabilities = [1 - p, p]

# Plotting the Bernoulli curve
plt.bar(x, probabilities, color=['blue', 'green'])
plt.xticks([0, 1], ['Failure (0)', 'Success (1)'])
plt.ylabel('Probability')
plt.title(f'Bernoulli Distribution (p = {p})')
plt.show()
```

### Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Example dataset
data = [
    ("Free money, Free money offer", "spam"),
    ("Congratulations, you won a lottery", "spam"),
    ("Call me when you can", "ham"),
    ("Meeting at 5 PM", "ham"),
    ("Urgent! Claim your prize", "spam"),
    ("Let's catch up tomorrow", "ham"),
]

# Separate the data into features and labels
texts, labels = zip(*data)
```

```python
import pandas as pd
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)  # Convert text to count vectors
df_X = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
df_X
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Loop through each column and plot a histogram
for column in df_X.columns:
    sns.histplot(df_X[column], label=column)  # Add kde=True for a smooth curve
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histograms of Each Column in df_X")
plt.legend(title="Columns")
plt.show()
```

```python
labels = labels # list(labels)
```

```python
# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=70,stratify=labels) # ,stratify=labels

# Create and train the model
model = MultinomialNB()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
model.score(X_test, y_test)
```

```python
import pandas as pd
import plotly.express as px
from sklearn.feature_extraction.text import CountVectorizer

# Sample text data
texts = ["This is a sample text", "This text is another example", "More sample text data"]

# Initialize CountVectorizer and transform texts
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)

# Convert the sparse matrix to a dense DataFrame with feature names as columns
df_X = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# Reshape the DataFrame to long format
df_melted = df_X.melt(var_name="Feature", value_name="Word Count")

# Plot using Plotly with hover information
fig = px.histogram(df_melted, x="Word Count", color="Feature", hover_data=["Feature"],
                   nbins=30, title="Histogram of Word Counts with Hover Labels")
fig.update_layout(xaxis_title="Word Count", yaxis_title="Frequency")
fig.show()
```
