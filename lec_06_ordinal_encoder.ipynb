```python
import pandas as pd
```

```python
ds = pd.DataFrame({"name":["s","m",'l','xl','m','s','l','xl','s','s','m']})
ds.head()
```

```python
order = [['s','m','l','xl']]
type(order)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=order)
```

```python
ds["orders"] = oe.fit_transform(ds)
```

```python
ds
```
