```python
import pandas as pd
```

```python
ds = pd.read_csv("SalaryData.csv")
ds.head(5)
```

```python
ds.isnull().sum()
```

```python
ds.dropna(inplace=True)
ds.isnull().sum()
```

```python
import seaborn as sns       
```

```python
sns.pairplot(ds)
```

```python
x = ds.iloc[:,:-1]
y = ds["Salary"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=.2, random_state=70)
```

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
```

```python
dtr = DecisionTreeRegressor()
```

```python
dtr.fit(x_train, y_train)
```

```python
dtr.score(x_test,y_test)*100
```

```python
import matplotlib.pyplot as plt
plot_tree(dtr)
plt.show()
```

```python
for i in range(1,20):
    dtr1 = DecisionTreeRegressor(max_depth=i)
    dtr1.fit(x_train, y_train)
    print(f"training score: {dtr1.score(x_train, y_train)} \t test score: {dtr1.score(x_test, y_test)} \t max_depth: {i}")
    
# we choose levels 5 because at this value both training and testing score almost same.
```

## post pruning

```python
dtr1 = DecisionTreeRegressor(max_depth=5)
dtr1.fit(x_train, y_train)
print(f"Training Score : {dtr1.score(x_train, y_train)*100}")
print(f"Test Score : {dtr1.score(x_test, y_test)*100}")
```

```python
plt.figure(figsize=(50,50))
plot_tree(dtr1)
plt.savefig("DTR.png")
plt.show()
```
