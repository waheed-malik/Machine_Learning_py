```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv("students_placement.csv")
ds.drop(columns=["profile_score"], inplace=True)
ds.head
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x=ds["cgpa"], y=ds["iq"], hue=ds["placed"])
plt.show()
# hue refers to the use of color to distinguish different categories or classes
# in a dataset.
```

```python
x = ds[["cgpa", "iq"]]
#x.ndim
y = ds["placed"]
y.ndim
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train, y_test = train_test_split(x,y, test_size=.20)
```

```python
x_train.shape,x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[9 ,101]])
```

```python
#%pip install mlxtend 
from mlxtend.plotting import plot_decision_regions
import numpy as np
```

```python
plt.figure(figsize=(5,3))
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=lr)
plt.show()
```
