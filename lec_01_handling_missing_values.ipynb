### Handling missing values
- keep in mind that if your dataset contain 50% null values then u dont need to use this dataset.
- but phr b if u want to use it bcz wo data set boht important hy to phr ap us column ko delete kr do jis
  me  missing vlaues 50% se ziyada hyn.

  Lakin ap kehty hyn k wo column bhi important hy to phr ap ne un rows ko delete kr dena hy. 

1 or kaam bhi kia ja sakta hy k hum delete na kry bal k un missing values ko khud fill kry.

But for now, hum delete krna seekhty hyn.

```python
import numpy as np
# Assuming ds is a pandas DataFrame
missing_data = ds.isnull().to_numpy()  # Convert the missing values to a NumPy array (True for missing, False otherwise)

# Create the heatmap using Matplotlib
plt.imshow(missing_data, cmap='Grays', aspect='auto')  # You can adjust cmap to any color map you like
plt.colorbar(label='Missing Data')  # Add a color bar
plt.xticks(ticks=np.arange(ds.shape[1]), labels=ds.columns, rotation=90)  # x-axis (columns)
plt.title('Missing Data Heatmap')
plt.show()
```

If your DataFrame ds has 5 columns named ['A', 'B', 'C', 'D', 'E'], 

np.arange(ds.shape[1]) will generate [0, 1, 2, 3, 4], which corresponds to the positions of each column.

labels=ds.columns will place the labels ['A', 'B', 'C', 'D', 'E'] at those positions.

rotation=90 will display those labels vertically.

```python
# suppose i want to delete Credit_History
ds.drop(columns = ['Credit_History'])

# ab ye jo result aya hy ye temporary hy. is ka impact hmary ds k ander jo data hy us pr nahi hwa. 
# agr hum chahty hyn k ds k dataset me changing ho to phir 1 property hy inplace usko true krna ho ga
ds.drop(columns = ['Credit_History'], inplace=True)
```
