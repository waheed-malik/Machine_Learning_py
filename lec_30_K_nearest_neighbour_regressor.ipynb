```python
import pandas as pd
```

```python

ds = pd.read_csv("SalaryData.csv")
ds.head(5)
```

```python
ds.describe()
```

```python
import seaborn as sns
sns.boxenplot(ds)
```

```python
ds.isnull().sum()
```

```python
ds.dropna(inplace=True)
ds.isnull().sum()
```

```python
sns.pairplot(ds)
```

```python
x = ds.iloc[:,:-1]
y = ds["Salary"]
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(x)
x = pd.DataFrame(ss.transform(x), columns=x.columns )
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
for i in range(1,20):
    knnr = KNeighborsRegressor(n_neighbors=i,p=1)
    knnr.fit(x_train, y_train)
    print(f"train score: {knnr.score(x_train,y_train)*100}, \t\t test score:{knnr.score(x_test,y_test)*100},\t i={i}")
    
```
