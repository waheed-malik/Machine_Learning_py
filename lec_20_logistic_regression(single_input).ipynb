### logistic regression 

-   is commonly used as a binary classifier, meaning it is used to predict 
        a binary outcome that has two possible values, such as "yes" or "no," "1" or "0," or "true" or "false."

-   For example, logistic regression might be used to predict whether:
        
        - A student will pass or fail an exam (two possible outcomes: pass/fail).
        - A customer will buy a product or not (buy/don't buy).            
        - A patient has a disease or not (disease/no disease).

-   use probabilities.

-   data nature must be linear (means data should be separable).

However, logistic regression can also be extended to handle more than two classes through multinomial logistic regression, but the basic form is a binary classifier.

#### Now work on a social networks ads data that tells a person
#### buy amazon subscription or not?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_csv("Social_Network_Ads.csv")
ds.drop(columns=["EstimatedSalary"], inplace=True)
```

```python
ds
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x=ds["Age"], y=ds["Purchased"])
plt.show()
```

```python
x = ds[["Age"]]
y = ds[["Purchased"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train, y_test = train_test_split(x,y,test_size=.20)
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[80]])
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x=ds["Age"], y=ds["Purchased"])
sns.lineplot(x=ds["Age"], y=lr.predict(x), color='red')
plt.show()
```
