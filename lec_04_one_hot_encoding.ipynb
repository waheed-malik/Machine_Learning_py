### get_dummies()

```python
# Using get_dummies to one hot encode 'color' column

dp = ds.copy()

dummies = pd.get_dummies(dp['Gender'], prefix='Gender', dummy_na=True) # dummy_na jitni unique values hon gi utny column bnae ga
#dummies2 = pd.get_dummies(dp['Married'], prefix='Married')
#dummies3 = pd.get_dummies(dp['Property_Area'], prefix='Property_Area')


dummies = dummies.astype(int)
#dummies2 = dummies2.astype(int)
#dummies3 = dummies3.astype(int)

#dp.drop('Gender', axis=1, inplace=True)
#dp.drop('Married', axis=1, inplace=True)

#encoded_df = pd.concat([dp,dummies3, dummies, dummies2,], axis=1)
encoded_df = pd.concat([dp,dummies], axis=1)

encoded_df
```

```python
# if columns in dataset is many then u can drop one column of each dummy columns.  agr ap k pass 1 column
# k 2 dummy column banty hyn to. otherwise nahi.

encoded_df.drop(['Gender_Female','Married_No'], axis=1, inplace=True)
encoded_df
```

```python
dd = ds
df = pd.DataFrame(dd)
# One hot encode 'Gender' column
gender_dummies = pd.get_dummies(df['Gender'], prefix='Gender')

# Convert True/False to 1/0 in 'Gender' dummy columns
gender_dummies = gender_dummies.astype(int)

# Drop the original 'Gender' column from the DataFrame
df.drop('Gender', axis=1, inplace=True)

# Concatenate the 'Gender' dummy columns back to the DataFrame
encoded_df = pd.concat([df, gender_dummies], axis=1)

encoded_df
```

## Using Sklearn library

```python
dt = ds[['Gender', 'Married']]
dt
```

```python
ss = pd.get_dummies(dt).info()
print(ss)
```

```python
# OneHotEncoder give result in form of sparse matrix (is me ziyada tar data 0 ki form me hota hy)
# and we can easily convert it into array using toarray() method

# fit pehly hmary data ko dekhta hy, us ko understand krta hy or phr sklearn ka model apply krta hy
# transform us data ko phr numerical me convert krta hy

#   %pip install scikit-learn
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()
# ohe.fit_transform(dt)
ar = ohe.fit_transform(dt).toarray()
ar
```

```python
after_label = pd.DataFrame(ar, columns=['Gender_Female', 'Gender_Male','Married_No', 'Married_Yes'] )
after_label.head()
```

```python
# hmary pass ye jo columns bany hyn in me 1 1 column directly bhi delete kr sakty hyn during 
# converting into dummies
# us k lye hmary pass 1 property hy drop ki wo use krty 

dtt = ds[['Gender', 'Married']]

ohe = OneHotEncoder(drop='first')
# ohe.fit_transform(dt)
arr = ohe.fit_transform(dtt).toarray()
arr
```

```python
after_label = pd.DataFrame(arr, columns=['Gender_Male','Married_Yes'] )
after_label.head()
```
