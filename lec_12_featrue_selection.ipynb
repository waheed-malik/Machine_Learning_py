#### Feature Selection

is technique k zarye hum apne data set se features selection krty hyn. ta k apne model ko train kr saky.

ya phr aise bhi keh sakty hyn k hum un-important columns/features ko apne dataset se remove krty hyn

is k lye 2 techniques hyn.

    - Forwarad Elimination

    - Backword Elimination

```python
import pandas as pd
```

```python
df = pd.read_csv("diabetes.csv")
df.head(5)
```

```python
x = df.iloc[: , :-1]
y = df["Outcome"]
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
# pip install mlxtend
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector
sf = SequentialFeatureSelector(lr,k_features=5, forward=True)
sf.fit(x,y)
```

```python
sf.k_feature_names_
```

```python
sf.k_score_
```

```python
# 5-features   ('Glucose'	 'BloodPressure'	'SkinThickness'	'BMI'	'Age') 0.7682794329853152
# 4-features   (('Glucose'	'BloodPressure'	'BMI'	'Age')	0.7682709447415329)
# 3-features    (('Glucose'	'BMI'	'Age')	0.7683048977166624)
# 2-features    (('Glucose'	'BMI')	0.7591206179441474)
# 1-features    (('Glucose'	)	0.7474747474747474)
```

```python
sf = SequentialFeatureSelector(lr,k_features=4, forward=True)
sf.fit(x,y)
sf.k_feature_names_, sf.k_score_
```

```python
sf = SequentialFeatureSelector(lr,k_features=3, forward=True)
sf.fit(x,y)
sf.k_feature_names_, sf.k_score_
```

```python
sf = SequentialFeatureSelector(lr,k_features=2, forward=True)
sf.fit(x,y)
sf.k_feature_names_, sf.k_score_
```

```python
sf = SequentialFeatureSelector(lr,k_features=1, forward=True)
sf.fit(x,y)
sf.k_feature_names_, sf.k_score_
```

### Backward Elimination technique

```python
sf = SequentialFeatureSelector(lr,k_features=5, forward=False)
sf.fit(x,y)
sf.k_feature_names_, sf.k_score_
```
